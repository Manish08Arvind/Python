# Pie Charts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
slices = [80,20]
label = ['Eighty','Twenty']

plt.pie(slices, labels=label)
plt.title('A simple pie chart')

plt.show()
```

**Wedge**

edge formatting

```python
slices = [80,20]
label = ['Eighty','Twenty']
color_list = ['skyblue','lightgreen']

plt.pie(slices, labels=label, colors=color_list,wedgeprops={'edgecolor':'black'})
plt.title('A simple pie chart')

plt.show()
```

```python
user_count = [59219, 55466, 47544, 36443, 35917]
lang_name = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(user_count,labels=lang_name,wedgeprops={'edgecolor':'black'})

plt.title('Programming languages by User proportion')

plt.show()
```

**Explode**

Pulls out the specified slice from the pie chart by the measure specified in the explode list. 

```python
user_count = [59219, 55466, 47544, 36443, 35917]
lang_name = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode_list = [0,0,0,0.2,0]

plt.pie(user_count,labels=lang_name,wedgeprops={'edgecolor':'black'}
        ,explode=explode_list)

plt.title('Programming languages by User proportion')

plt.show()
```

**Shadow**

```python
user_count = [59219, 55466, 47544, 36443, 35917]
lang_name = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode_list = [0,0,0,0.2,0]

plt.pie(user_count,labels=lang_name,wedgeprops={'edgecolor':'black'}
        ,explode=explode_list,shadow=True)

plt.title('Programming languages by User proportion')

plt.show()
```

**startangle**

The default startangle is 0 and specifyin it to 90 will rotate the chart counter-clockwise by 90 degrees.

```python
user_count = [59219, 55466, 47544, 36443, 35917]
lang_name = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode_list = [0,0,0,0.2,0]

plt.pie(user_count,labels=lang_name,startangle=90,
        explode=explode_list,shadow=True,wedgeprops={'edgecolor':'black'})

plt.title('Programming languages by User proportion')

plt.show()
```

**autopct**

Autopct for percentage display.

```python
user_count = [59219, 55466, 47544, 36443, 35917]
lang_name = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode_list = [0,0,0,0.2,0]

plt.pie(user_count,labels=lang_name,startangle=90,explode=explode_list,
        shadow=True,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})

plt.title('Programming languages by User proportion')

plt.show()
```
