```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from warnings import filterwarnings

filterwarnings('ignore')
```

```python
# data read

data = pd.read_csv('data1.csv')
data.head()
```

```python
ages = data['Age']
coders_sal = data['All_Devs']
python = data['Python']
javascript = data['JavaScript']
```

```python
# Simple line plot

plt.plot(ages,coders_sal,color='#444444',linestyle='--', label='All Coders')
plt.plot(ages,python,label='Python')

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Age (in years)')
plt.ylabel('Median Salary (in $)')

plt.show()
```

Fill between

The fill_between() function creates a polygon that fills the area between the two line. The color of the polygon can be changed by setting the color argument. 

alpha

To handle the color shade

```python
# Fill Area on Line Plots
plt.plot(ages,coders_sal,color='#444444',linestyle='--', label='All Coders')
plt.plot(ages,python,label='Python')

plt.fill_between(ages,python,alpha=0.4)

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Age (in years)')
plt.ylabel('Median Salary (in $)')

plt.show()
```

filling between overall median and python salary values

```python
plt.plot(ages,coders_sal,color='#444444',linestyle='--', label='All Coders')
plt.plot(ages,python,label='Python')

overall_median = 57287
plt.fill_between(ages,python,y2=overall_median,alpha=0.4)

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Age (in years)')
plt.ylabel('Median Salary (in $)')

plt.show()
```

Filling the area with only python salaries are greater than overal median 

```python
plt.plot(ages,coders_sal,color='#444444',linestyle='--', label='All Coders')
plt.plot(ages,python,label='Python')

overall_median = 57287
plt.fill_between(ages,python,y2=overall_median,where=(python>overall_median),
                 interpolate=True,alpha=0.4)

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Age (in years)')
plt.ylabel('Median Salary (in $)')

plt.show()
```

Second fill for the below area

```python
plt.plot(ages,coders_sal,color='#444444',linestyle='--', label='All Coders')
plt.plot(ages,python,label='Python')

overall_median = 57287
plt.fill_between(ages,python,y2=overall_median,where=(python>overall_median),
                 interpolate=True,alpha=0.4)

plt.fill_between(ages,python,y2=overall_median,where=(python<=overall_median),
                 interpolate=True,color='red',alpha=0.4)

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Age (in years)')
plt.ylabel('Median Salary (in $)')

plt.show()
```

fill between all coders and python area alone

```python
plt.plot(ages,coders_sal,color='#444444',linestyle='--', label='All Coders')
plt.plot(ages,python,label='Python')

overall_median = 57287
plt.fill_between(ages,python,y2=coders_sal,where=(python>coders_sal),
                 interpolate=True,alpha=0.4,label='Above Avg.')

plt.fill_between(ages,python,y2=coders_sal,where=(python<=coders_sal),
                 interpolate=True,color='red',alpha=0.4,label='Below Avg.')

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Age (in years)')
plt.ylabel('Median Salary (in $)')

plt.show()
```
