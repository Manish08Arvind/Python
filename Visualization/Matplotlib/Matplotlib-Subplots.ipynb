```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv('data1.csv')
df
```

```python
ages = df['Age']
all_dev_sal =df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']
```

```python
plt.plot(ages, py_sal, label = 'Python')
plt.plot(ages, js_sal, label = 'JavaScript')
plt.plot(ages, all_dev_sal, color='#444444', linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (in $) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (in $)')

plt.show()
```

Subplots

```python
# Inital and simple sub plot
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(ages, all_dev_sal, color='#444444', linestyle='--', label='All Devs')

ax2.plot(ages, py_sal, label = 'Python')
ax2.plot(ages, js_sal, label = 'JavaScript')

ax1.legend()
ax1.set_title('Median Salary (in $) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (in $)')

ax2.legend()
ax2.set_title('Median Salary (in $) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (in $)')

plt.show()
```
