```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

Basic Scatter Plot

```python
plt.scatter(x,y)
```

Size of dots, color, markers, edgecolor, linewidth. alpha

```python
plt.scatter(x,y,s=100,c='violet',marker='.',edgecolor='black',linewidth=1, alpha=0.9)
```

Multiple colors

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y,s=100,c=colors,edgecolor='black',linewidth=1,alpha=0.9)
```

cmap - To set to specify color of different shades

```python
plt.scatter(x,y,s=100,c=colors,cmap='Greens',edgecolor='black',linewidth=1,alpha=0.9)
```

size of the dot

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x,y,s=sizes,c=colors,cmap='Greens',edgecolor='black',linewidth=1,alpha=0.9)

color_bar = plt.colorbar()
color_bar.set_label('Satisfaction Level')
```

The dark circles indicate highest level of satisfaciton and the lighter ones denote less satisfaction. 

```python
df = pd.read_csv('youtube_api_sample_data.csv')
df
```

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, edgecolor='black',linewidth=1,alpha=0.8)

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```

log scale conversion

```python
plt.scatter(view_count, likes, edgecolor='black',linewidth=1,alpha=0.8)

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```

color map scale using 'ratio'

```python
plt.scatter(view_count,likes,c=ratio,cmap='BuPu',edgecolor='black',
            linewidth=1,alpha=0.8)

color_bar = plt.colorbar()
color_bar.set_label('Like\Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.show()
```
