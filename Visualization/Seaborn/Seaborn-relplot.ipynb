```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings('ignore')
```

Seaborn is used for data visualization and based on Matplotlib. 

It allows the creation of statistical graphs.

**Functionalities**

* Allows comparison between multiple variables.

* Supports multi-plot grids.

* Available univariate and bivariate visualizations.

* Availability of different color palettes.

* Estimates and plots linear regression automatically.

**Advantages of Seaborn over Matplotlib**

* Matplotlib settings are difficult to figure out. Seaborn comes with numerous customized themes and high-level interfaces.

* Matplotlib doesn't serve well when it comes to dealing with DataFrames while Seaborn functions actually work on DataFrames.

**Functions of Seaborn**

>**Visualizing statistical relationships** - Process of understanding relationships  between variables of a dataset.

>**Plotting with categorical data** - Main variable is further divided into discrete groups.

>**Visualizing the distribution of a dataset** - Understanding datasets with context to being univariate or bivariate.

```python
df = sns.load_dataset('flights')
df
```

## relplot

**The relplot function is used to create a relational plot, which is a high-level interface for creating various types of plots that show the relationship between two variables.**

**It is similar to the _'scatterplot'_ and _'lineplot'_ functions, but it provides more flexibility and can create multiple types of relational plots.**

**Key Parameters of relplot**

> 'x' and 'y' : These parameters specify the variables you want to plot on the x and y axes, respectively.

> 'data' : This parameter is used to specify the DataFrame or data source from which the variable are drawn.

> 'kind' : This parameter determines the type of plot to create. It can be one of the following:
>>"scatter"(default): Creates a scatter plot.<br>
>>"lineplot": Creates a line plot.

### Scatter Plot

```python
# relplot - scatter  plot(default)

sns.relplot(x='passengers',y='month',data=df)
```

```python
# specifying hue semantic

sns.relplot(data=df,x='passengers',y='month',hue='year')
```

```python
# Scatter plotting individually by 'year'
sns.relplot(data=df,x='passengers',y='month',hue='year',col='year',col_wrap=3,palette='hls')
```

### Line Plot

```python
df1 = sns.load_dataset('tips')
df1
```
