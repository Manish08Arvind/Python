```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
df
```

### Hist Plot

```python
sns.displot(data=df,x='total_bill',hue='sex',multiple='stack')
```

Decoding the stacked histogram interpretation. From the above we'll just pick the bin 5-10 and in that the orange portion covers from 0 to 5 and the rest from 5 upto approx 15 is covered by blue. Thereby, in this bin there are 5 females and males = 15-5 = 10(blue region).

```python
# Data fetch for the above graph and detailing:
df[(df['total_bill']>=5) & (df['total_bill']<=10)]['sex'].value_counts()
```

### Kde Plot

```python
sns.displot(data=df,x='total_bill',hue='time',kind='kde')
```

### Rug Plot

It plots marginal distributions by drawing ticks along the x and y axis. Mostly used with other plots to provide a visual representation of the data points and their density.

This plot is intended to show location of individual observation in a unobtrusive way.

```python
sns.displot(data=df,x='total_bill',kde=True,rug=True,rug_kws={'height':0.05})
```

### Ecdf Plot

**Empirical Cumulative Distribution Function(ECDF)** plot provides a visual representation of the cumulative distribution of a single quantitative variable. It shows how the data points are distributed across the range of values and allows to understand the cumulative probability of observing a value less than or equal to a praticular point on the x-axis.

```python
sns.displot(data=df,x='total_bill',kind='ecdf')
```

From the above graph we can interpret that 80% of the total bill amounts are less than 25(approx.).

```python
sns.displot(data=df,x='total_bill',hue='time',kind='ecdf')
```
