## catplot

The **catplot** function is used to create categorical plots, which are specialized plots for visualizing the ditribution of a variable across multiple categories or levels of another variable.

It is a high-level interface that allows you to create different types of categorical plots, such as strip plots, swarm plots, bar plots and more depending on the value of the _'kind'_ parameter.

**Key Parameters**

* **'x' and 'y'**: It specifies the variables to be plotted on the x and y axis. The chocie of 'x' and 'y' depends on the specific kind of categorical plot you want to create.
    
* **data**: This parameter is used to specify the DataFrame or data source from which the variables are drawn.
    
* **kind**: This parameter determines the type of categorical plot to create. It can be on of the follwoing:
>"strip"(default): Creates a strip plot<br>
>"swarm"<br>
>"box"<br>
>"violin"<br>
>"boxen"<br>
>"point"<br>
>"bar"<br>
>"count"<br>

The choice of **'kind'** will determine the specific visualization used to represent the data in a categorical context.

```python
df = sns.load_dataset('tips')
df
```

### Strip Plot

```python
sns.catplot(data=df,x='day',y='total_bill',hue='day')
```

### Violin Plot 

```python
sns.catplot(data=df,x='day',y='total_bill',kind='violin')
```

### Boxen Plot

```python
sns.catplot(data=df,x='day',y='total_bill',kind='boxen')
```

The quartiles start from the middle q2 as 50th percentile i.e. 1/2, then q1 = 1/4 then q3 = 1-1/4, right side of q3 goes as follows 1-1/8, 1-1/16. left side 1/8, 1/16

### Swarm Plot

```python
sns.swarmplot(data=df,x='day',y='total_bill',hue='day')
```

Swarm plots may resemble strip plots in character of plotting data points along the x-axis or categorical axis to their respecitve values in the y-axis.

**However, in swarm plots similar data points do not overlap rather they spread horizontally which is useful when handling large dataset**

### Point Plot

```python
sns.pointplot(data=df,x='day',y='total_bill')
```

The line in the above point plot is the confidence interval _(by default 95%, can be changed accordingly)_ and the marker or the dark blue dot is the mean of the respective day's total_bill amount

```python
# changing to 60% confidence interval and we can see the change in range

sns.pointplot(data=df,x='day',y='total_bill',ci=60)
```

### Bar Plot

```python
sns.barplot(data=df,x='day',y='total_bill')
```
