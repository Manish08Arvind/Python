## Ranking

#### Different methods in ranking

Group of records have same value....oh..How do we rank it?

method to the rescue with the following options:
> average : Average rank of the group

> min : Lowest rank in the group

> max : Highest rank in the group

> first : Ranks assigned in order they appear in the array

> dense : like 'min', but rank always increases by 1 between groups

```python
# Rank based on volume

# Min

df['Vol_Rank_min']=df['Vol'].rank(axis=0,method='min',ascending=True)
df
```

```python
# Dense

df['Vol_Rank_Dense'] = df['Vol'].rank(axis=0,method='dense',ascending=True)
df
```

```python
# first

df['Vol_Rank_first'] = df['Vol'].rank(axis=0,method='first',ascending=True)
df
```

```python
# average (Not a preferred one in ranking)

df['Vol_Rank_average'] = df['Vol'].rank(axis=0,method='average',ascending=True)
df
```

```python
# max

df['Vol_Rank_max']=df['Vol'].rank(axis=0,method='max',ascending=True)
df
```

Intersection created with "inner" results out with only common columns

inner

```python
pd.merge(left = df21, right = df22, on = 'HQ', how = 'inner')
```

outer

```python
pd.merge(left = df21, right = df22, on = 'HQ', how = 'outer')
```

Indicator

```python
pd.merge(left = df21, right = df22, on='HQ', how='outer',indicator=True)
```

Left

```python
pd.merge(left = df21, right = df22, on='HQ', how='left',indicator=True)
```

Right

```python
pd.merge(left = df21, right = df22, on='HQ', how='right',indicator=True)
```

cross join

```python
pd.merge(left = df21, right = df22,how='cross',indicator=True)
```

## Join

```python
df21
```

```python
df22
```

```python
# By default a left join occurs and is joined with index as the key

df21.join(df22,lsuffix = '_left',rsuffix = '_right')
```

```python
df21.join(df22,how='right',lsuffix = '_left',rsuffix = '_right')

# Here, we can observe a right join has happened resulting in the presence of Bombay_Finance with NaNs in the left 
# and NCR_Leagles not appearing from the left dataframe(df21)
```

```python
# Common index are returned from both dataframes

df21.join(df22,how='inner',lsuffix = '_left',rsuffix = '_right')
```

```python
# All indexs and their respective values are populated in the resultant dataframe

df21.join(df22,how='outer',lsuffix = '_left',rsuffix = '_right')
```

```python
# Cross join

df21.join(df22,how='cross',lsuffix = '_left',rsuffix = '_right')
```

```python
# Trying to use any other column to join the dataframes

# Since join happens at index level a set index needs to be passed to the joining dataframe for any column other than 
# default index of the dataframe
df21.join(df22.set_index('HQ'),how='left',on='HQ')
```
