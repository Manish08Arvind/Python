### Difference between Pivot table and Pivot

**Pivot table:** It summarizes data by grouping into categories and calculating various aggregate statistics.
    
**Pivot:** The pivot method is used to reshape data by specifying the index, columns and values of a dataframe.

**Pivot table**

The pivot table method takes several arguments, including values, index, columns, aggfunc, and fill_value.

The values argument specifies the column that will be used as the values in the pivot table, while index and columns arguments specify the columns that will be used as the rows and columns of the pivot table.

The aggfunc argument specifies the aggregation function to be used, such as sum, mean or count.

The fill_value argument specifies a value to replace missing values in the pivot table.

**Pivot**

The pivot method takes three main arguments: index, columns and values.
    
The index arguments specifies the columns that will be used as the new index, while the columns argument specifies the columns that will be used as the new columns.

The values argument specifies the columns that will be used as the new values.

The pivot method will then reshape the data in the dataframe to match the specified index, columns and values.

## Stacking and Unstacking

#### Stack

> It changes the wide table to a long table.

> Returns a reshaped dataframe of Multi-index.

> The new inner most levels are created by pivoting the columns of current dataframe.

```python
df1
```

```python
df1.stack() 
# level is -1 by default - level is used to stack from columns onto index
```

```python
pd.DataFrame(df1.stack(),columns = ['Metrics'])
```

#### Unstack

> unstack has columns as index labels in the first

```python
df1.unstack()
```

```python
pd.DataFrame(df1.unstack(),columns=['Metrics'])
```

```python
# If we chainn the unstack method 4 time we'll get the original dataframe back
for i in range(4):
    df1 = df1.unstack()
    
df1
```
